```python
plt.xlabel("Quilates")
plt.ylabel("Precio")
plt.scatter(diamantes.quilate, diamantes.precio)
plt.show()
```

```python
# hay mucho ruido
# si filtramos un poco
diamantes_premium = diamantes[diamantes["corte"] == "Premium"]
```

```python
plt.title("Corte: Premium")
plt.scatter(diamantes_premium.quilate, diamantes_premium.precio)
plt.show()
```

```python
diamantes_premium_E_SI1 = diamantes[(diamantes["corte"] == "Premium") &
                                     (diamantes["color"] == "E") &
                                     (diamantes["claridad"] == "SI1")
                                     ]
```

```python
plt.title("Corte: Premium; Color: E; Claridad: SI1")
plt.scatter(diamantes_premium_E_SI1.quilate, diamantes_premium_E_SI1.precio)
plt.show()
```
